First Seen in the Cat Herding Journal: 

It is no secret to the cat herding community that the easiest cats to herd are old ones. We are talking 15+ years old. Once they hit double digits it becomes much easier to coerce a cat to doing what you want. Kittens have also found to be fairly good for cat herding, provided you are diligent with training and do not let them roam outside of the house, except for herding training. 

This being said, age is but a number. While these ages have been proven to yield easier results, it does not mean they will yield the best results. An excellent cat herding will be able to herd no matter the age (or breed!)
